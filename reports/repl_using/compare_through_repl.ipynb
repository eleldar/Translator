# Разница в переводах при использовании скриптов repl
> скрипты repl могут использоваться вместе с предобученными моделях (preprocess.sh), загруженных с сайта huggingface.co; их использование способствуют более качественному переводу.

> Ниже рассмотрена разница в переводах с английского на русский язык (из 68 предложений 11 имеют несущественную, но местами заментую разницу, например: "документом о Кремлях").

```python
with open("no_repl.txt", encoding='utf-8', newline='') as f:
    no = f.readlines()
no = [i.strip() for i in no]
```

```python
with open("yes_repl.txt", encoding='utf-8', newline='') as f:
    yes = f.readlines()
yes = [i.strip() for i in yes]
```

```python
f'всего предложений {len(no)}, {len(yes)}'
```

```python
diff_sent = 0
for n, y in zip(no, yes):
    if n != y:
        print(n, y, sep='\n --(c repl)--> \n')
        print('*' * 100)
        diff_sent += 1
f'{diff_sent} предложений имеют различия'
```
