```python
def get_list(path_file):
    with open(path_file, encoding='utf-8') as f:
        return f.readlines()


paths = ['flores101_dataset/dev', 'flores101_dataset/devtest']
dev_names = set([i.split('.')[0] for i in os.listdir(paths[0])])
devtest_names = set([i.split('.')[0] for i in os.listdir(paths[1])])
names = dev_names | devtest_names
result = dict.fromkeys(names, [])

for path in paths:
    for file in os.listdir(path):
        path_file = f'{path}/{file}'
        name = file.split('.')[0]
        tmp = result[name][:]
        tmp.extend(get_list(path_file))
        result[name] = tmp[:]
```
